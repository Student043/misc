```python
import autosklearn
import sklearn.datasets
import numpy as np


DATASETS = (
    sklearn.datasets.load_digits(),
    sklearn.datasets.load_iris(),
)
    

def split_data(dataset):
    X = dataset.data
    y = dataset.target
    indices = np.arange(X.shape[0])
    np.random.shuffle(indices)
    X = X[indices]
    y = y[indices]

    train_amount = int(X.shape[0] * 0.9)
    X_train = X[:train_amount]
    y_train = y[:train_amount]
    X_test = X[train_amount:]
    y_test = y[train_amount:]
    return (X_train, y_train, X_test, y_test)


def run(dataset):
    name = dataset.DESCR.splitlines()[0].strip()
    (X_train, y_train, X_test, y_test) = split_data(dataset)

    clf = autosklearn.AutoSklearnClassifier(time_left_for_this_task=360, per_run_time_limit=36)
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    return (name, score)


result = map(run, DATASETS)
```
